```python
x1=np.random.normal(1,1, size=(100,1))
x2=np.random.normal(-2,4, size=(100,1))
x=np.concatenate((x1, x2), axis=1)      #axis:  ( 1: 가로 , 0:세로)
x
```

```python
#산점도 그래프 - 상관관계를 보여주기 위한 그래프
plt.scatter(x1, x2)
```

```python
df=pd.DataFrame(x, columns=['x1','x2'])
#상관 계수 출력 : 1에 가까울 수록 상관관계가 높음!!
df['x1'].corr(df['x2'])
```
