# 5. cat plot
- 범주를 나누어서 그려줌
- 참조 사이트: https://seaborn.pydata.org/generated/seaborn.catplot.html
- kind (swarm, violin, box, boxen, point, bar, count)


```python
sns.catplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex',inner='stick', split=True)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='point')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, hue='sex', kind='bar')
```

```python
sns.catplot(x='day', data=tips, hue='sex', kind='count')
```

```python
# ex) titanic 데이터셋을 로드하여 x=성별, y=survived, hue=class를 사용하여 bar chart를 작성하시오.
#       남녀, 클래스 별 생존자 확인차트

titanic=sns.load_dataset('titanic')
print(sns.catplot(x='sex', y='survived', data=titanic, hue='class', kind='bar'))
print(sns.catplot(x='class', y='survived', data=titanic, hue='sex', kind='bar'))
print(sns.catplot(x='class', y='survived', data=titanic, col='sex', kind='bar'))
```
