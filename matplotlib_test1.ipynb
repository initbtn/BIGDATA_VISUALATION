```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

#파이차트 한글 사용 설정 방법
from matplotlib import rc, font_manager
fontname="C:/Windows/Fonts/H2GTRE.TTF"
#C:/Windows/Fonts/BMYEONSUNG_ttf.ttf
font_name=font_manager.FontProperties(fname="C:/Windows/Fonts/H2GTRE.TTF").get_name()
rc('font',family=font_name)
```

```python
filename='data/temperature_2014.csv'
df=pd.read_csv(filename)
df.info()
```

```python
df['PST']=pd.to_datetime(df['PST'])
df.info()
```

```python
#캔버스 준비
fig=plt.figure(dpi=128, figsize=(10,6))

plt.plot(df['PST'],df['Max TemperatureF'], c='red', alpha=0.5)
plt.plot(df['PST'],df['Min TemperatureF'], c='blue', alpha=0.5)
#색 칠하기
plt.fill_between(df['PST'], df['Max TemperatureF'], df['Min TemperatureF'],
                 facecolor='blue', alpha= 0.1)
#
plt.title("온도 통계", fontsize=20)
plt.xlabel("기간",fontsize=16)
plt.ylabel("온도",fontsize=16)
plt.tick_params(axis='both',labelsize=16)
plt.savefig('data/chart.png', dpi=400, bbox_inches='tight')
plt.show()
```
