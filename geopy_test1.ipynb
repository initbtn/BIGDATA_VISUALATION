```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
df=pd.read_csv('data/CoffeeBean.csv')
df.address
```

```python
geolocoder=Nominatim(user_agent='South Korea')      #객체 생성
```

```python
a=['서울특별시 강남구 학동로 211', '서울특별시 강남구 광평로 280', '서울특별시 강남구 논현로 566','부산 사하구 사하로141번길 63-2','부산 부산진구 중앙대로 708']
print(enumerate(a))
for i, loc in enumerate(a):
    geo=geolocoder.geocode(loc)
    crd=(geo.latitude,geo.longitude)
    print(geo, crd)
```
