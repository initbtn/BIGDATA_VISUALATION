# 2. relplot
- 두 columns 간의 상관 관계를 보여줌

```python
#sns.relplot(x='x축으로 사용할 필드', y='y축으로 사용할 필드',data='dataFrame')
sns.relplot(x='total_bill',y='tip', data=tips)
print(tips['total_bill'].corr(tips['tip']))
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='day')
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, style='smoker', hue='size')
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', col='day', row='sex', data=tips)
```

x:total, y: tip, row: payment, col: color

```python
sns.get_dataset_names()
taxis=sns.load_dataset('taxis')
sns.relplot(x='total', y='tip', col='color', row='payment', data=taxis)
```

```python
df=pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df
```

```python
#중복값 제거
fmri=sns.load_dataset('fmri')
fmri
```

```python
# sns.distplot(): 히스토그램, 라인그래프, kde 그래프
x=np.random.randn(100)
```

```python
sns.distplot(x, rug=True, hist=True, kde=True)
```

```python
sns.distplot(x, rug=True, hist=True, kde=False)
```

```python
sns.histplot(data=x, kde=True)
sns.rugplot(x=x)
```

```python
sns.histplot(data=tips.total_bill, kde=True)
sns.rugplot(x=tips.total_bill)
```
