### 1. 데이터 불러오기 

```python
cb=pd.read_csv('data/CoffeeBean.csv')
cb
```

### 2. 주소명 컬럼을 추출해서 [시도, 구군] 리스트 형태 데이터로 만들기 

```python
add=pd.DataFrame(cb.address.apply(lambda v: v.split()[:2]).to_list(),columns=['시도','구군'])
add
```

### 3. 시도 필드의 중복값 제거 

```python
add.시도.unique()
```

### 4. 서울, 서울시 값을 '서울특별시'로 변경

```python
addr_aliases={'서울':'서울특별시', '서울시':'서울특별시'}
add['시도']=add['시도'].apply(lambda v: addr_aliases.get(v, v))
add['시도'].unique()
```

### 5. add에서 특정 데이터만 추출하기

```python
addr2=add.loc[add['시도']=='서울특별시',:]
addr2
```

### 6. 구군 필드로 그룹화(groupby), count

```python
addr_group=addr2.groupby('구군')['시도'].count()
addr_group
```

### 7. 구군 경계로 카운트 데이터 히트맵 시각화하기 

```python
m.choropleth(geo_data=seoul_geo,
                data=addr_group,
                fill_color='YlOrRd',
                fill_opacity=0.7,
                line_opacity=0.2,
                key_on='properties.name',
                legend_name='지역구별 커피 업종수')
m
```
