```python
from folium import plugins
n=100
data=np.array([np.random.uniform(low=35,high=38, size=n),np.random.uniform(low=126, high=129, size=n)])
data
```

```python
data=data.T
```

```python
popups=[str(i) for i in range(n)]
popups
```

```python
m=folium.Map(location=[latitude,longitude],zoom_start=8)
plugins.MarkerCluster(data,popups=popups).add_to(m)
m.save('data/cluster.html')
m
```

```python
import pandas as pd
df=pd.read_csv('data/CCTV_20190917.csv')
df1=df.loc[df['카메라대수']>0,['소재지도로명주소','위도','경도']]

popups=df1['소재지도로명주소'].to_list()
popups
```

```python
lat=df1['위도'].to_list()
log=df1['경도'].to_list()

data=np.array([lat,log]).T
```

```python
data
```

```python
m=folium.Map(location=[35.1803305, 129.0516257],zoom_start=15)
plugins.MarkerCluster(data,popups=popups).add_to(m)
m.save('data/cluster.html')
m
```

```python
import requests
import json
r=requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c=r.content
seoul_geo=json.loads(c)
seoul_geo
```

```python
m=folium.Map(location=[latitude, longitude], zoom_start=11)
folium.GeoJson(seoul_geo, name='지역구').add_to(m)
m
```
